## **Pelatihan Model**

```python
# ==================
# 1. Import & Setup
# ==================
import torch
import torch.nn as nn

from detectron2.config import get_cfg
from detectron2.engine import DefaultTrainer
from detectron2.modeling import BACKBONE_REGISTRY, Backbone, build_backbone
from detectron2.layers import Conv2d, ShapeSpec
# Import FPN & LastLevelMaxPool dari submodule fpn
from detectron2.modeling.backbone.fpn import FPN, LastLevelMaxPool

from detectron2.checkpoint import DetectionCheckpointer
from detectron2.data import build_detection_test_loader
from detectron2.evaluation import inference_on_dataset, COCOEvaluator
from detectron2.utils.logger import setup_logger

setup_logger()

# ==========================================
# 2. Registrasi Dataset (Opsional)
# ==========================================
# Jika diperlukan, Anda bisa mendaftarkan dataset Anda. Contoh:
# from detectron2.data.datasets import register_coco_instances
# register_coco_instances("my_dataset_train", {}, "path/to/train.json", "path/to/images_train")
# register_coco_instances("my_dataset_val", {}, "path/to/val.json", "path/to/images_val")

# ======================================================
# 3. Definisikan MobileNetV3 Small sebagai custom backbone (bottom-up)
# ======================================================
@BACKBONE_REGISTRY.register()
def build_mobilenetv3_small_backbone(cfg, input_shape: ShapeSpec):
    import torchvision.models as models

    # Gunakan pre-trained weights untuk MobileNetV3 Small
    weights = models.MobileNet_V3_Small_Weights.IMAGENET1K_V1
    model = models.mobilenet_v3_small(weights=weights)

    # Pembagian stage sesuai downsampling yang diharapkan:
    # stage1: features[:2]  -> 16 channel, effective stride ~4
    # stage2: features[2:4]  -> 24 channel, effective stride ~8
    # stage3: features[4:6]  -> 40 channel, effective stride ~16
    # stage4: features[6:9]  -> 48 channel, tetap dengan stride ~16
    # stage5: features[9:]   -> 576 channel, effective stride 32
    stage1 = nn.Sequential(*model.features[:2])
    stage2 = nn.Sequential(*model.features[2:4])
    stage3 = nn.Sequential(*model.features[4:6])
    stage4 = nn.Sequential(*model.features[6:9])
    stage5 = nn.Sequential(*model.features[9:])

    class MobileNetV3Backbone(Backbone):
        def __init__(self):
            super().__init__()
            self.stage1 = stage1
            self.stage2 = stage2
            self.stage3 = stage3
            self.stage4 = stage4
            self.stage5 = stage5

            # (Opsional) Modul peningkatan untuk fitur resolusi tinggi (dari stage1)
            self.extra_p2 = nn.Sequential(
                Conv2d(16, 16, kernel_size=3, padding=1, bias=False),
                nn.BatchNorm2d(16),
                nn.ReLU()
            )

            # Proyeksi ke 256 channel untuk FPN:
            # p2: dari stage1 (stride 4)
            # p3: dari stage2 (stride 8)
            # p4: dari stage3 & stage4 (stride 16)
            # p5: dari stage5 (stride 32)
            self.conv_p2 = Conv2d(16, 256, kernel_size=1, bias=False)
            self.conv_p3 = Conv2d(24, 256, kernel_size=1, bias=False)
            self.conv_p4 = Conv2d(48, 256, kernel_size=1, bias=False)
            self.conv_p5 = Conv2d(576, 256, kernel_size=1, bias=False)

        def forward(self, x):
            # Stage 1: Output stride ~4
            x = self.stage1(x)
            x_enhanced = self.extra_p2(x)
            p2 = self.conv_p2(x_enhanced)

            # Stage 2: Output stride ~8
            x = self.stage2(x)
            p3 = self.conv_p3(x)

            # Stage 3 & 4:
            # Stage3 menghasilkan 40 channel dengan stride ~16, kemudian diproses di stage4 untuk menghasilkan 48 channel
            x = self.stage3(x)
            x = self.stage4(x)
            p4 = self.conv_p4(x)

            # Stage 5: Output stride 32
            x = self.stage5(x)
            p5 = self.conv_p5(x)

            return {"p2": p2, "p3": p3, "p4": p4, "p5": p5}

        def output_shape(self):
            return {
                "p2": ShapeSpec(channels=256, stride=4),
                "p3": ShapeSpec(channels=256, stride=8),
                "p4": ShapeSpec(channels=256, stride=16),
                "p5": ShapeSpec(channels=256, stride=32),
            }

    return MobileNetV3Backbone()

# ======================================================
# 4. Bungkus backbone di atas dengan FPN
# ======================================================
@BACKBONE_REGISTRY.register()
def build_mobilenetv3_small_fpn_backbone(cfg, input_shape):
    """
    Membungkus MobileNetV3Small backbone dengan FPN.
    Menggunakan output fitur: p2, p3, p4, p5 → FPN akan menghasilkan p2, p3, p4, p5, dan top block menghasilkan p6.
    """
    bottom_up = build_mobilenetv3_small_backbone(cfg, input_shape)
    fpn = FPN(
        bottom_up=bottom_up,
        in_features=["p2", "p3", "p4", "p5"],
        out_channels=256,
        norm=cfg.MODEL.FPN.NORM,
        top_block=LastLevelMaxPool(),  # Top block akan menghasilkan fitur "p6"
        square_pad=cfg.MODEL.FPN.SQUARE_PAD
    )
    return fpn

# ==========================================
# 5. Configurasi
# ==========================================
cfg = get_cfg()
cfg.merge_from_file("detectron2/configs/COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml")

# Gunakan custom backbone MobileNetV3Small dengan FPN
cfg.MODEL.BACKBONE.NAME = "build_mobilenetv3_small_fpn_backbone"
cfg.MODEL.FPN.IN_FEATURES = ["p2", "p3", "p4", "p5"]

# FPN output akan memiliki fitur: p2, p3, p4, p5, dan p6 (dari top block)
cfg.MODEL.RPN.IN_FEATURES = ["p2", "p3", "p4", "p5", "p6"]
cfg.MODEL.ROI_HEADS.IN_FEATURES = ["p2", "p3", "p4", "p5", "p6"]

# Sesuaikan anchor sizes (misal: level p2 mendapatkan anchor terkecil, 16 piksel)
cfg.MODEL.ANCHOR_GENERATOR.SIZES = [[16], [32], [64], [128], [256]]

cfg.MODEL.ROI_HEADS.NUM_CLASSES = 4
cfg.MODEL.WEIGHTS = ""
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
cfg.OUTPUT_DIR = "./output_mbv3_fpn"

cfg.DATASETS.TRAIN = ("my_dataset_train",)
cfg.DATASETS.TEST = ("my_dataset_val",)
cfg.DATALOADER.NUM_WORKERS = 2

cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.002
cfg.SOLVER.MAX_ITER = 20000
cfg.SOLVER.STEPS = (12000, 16000)
cfg.SOLVER.GAMMA = 0.1
cfg.SOLVER.WARMUP_ITERS = 1000
cfg.SOLVER.WARMUP_FACTOR = 1.0 / 1000
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128

cfg.MODEL.FPN.SQUARE_PAD = 0

cfg.SOLVER.OPTIMIZER = "SGD"
cfg.SOLVER.MOMENTUM = 0.9
cfg.SOLVER.WEIGHT_DECAY = 1e-4

# =========================================
# 6. Trainer
# =========================================
class Trainer(DefaultTrainer):
    @classmethod
    def build_evaluator(cls, cfg, dataset_name):
         return COCOEvaluator(dataset_name, cfg, False, output_dir=cfg.OUTPUT_DIR)

trainer = Trainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()

# =========================================
# 7. Evaluasi
# =========================================
val_loader = build_detection_test_loader(cfg, "my_dataset_val")
evaluator = Trainer.build_evaluator(cfg, "my_dataset_val")
results = inference_on_dataset(trainer.model, val_loader, evaluator)
print(results)
```

```python
import os
import json
import cv2
import numpy as np
import matplotlib.pyplot as plt

from detectron2.config import get_cfg
from detectron2.engine import DefaultPredictor
from detectron2.data import MetadataCatalog, DatasetCatalog
from detectron2.utils.logger import setup_logger

import torch
from detectron2.structures import Boxes, pairwise_iou
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

setup_logger()

# =========================================
# 1. Plot Training Curve dari metrics.json
# =========================================

def plot_training_curve(metrics_json_path="/content/output_mbv3_fpn/metrics.json"):
    """
    Membaca file metrics.json hasil training Detectron2 dan menampilkan grafik
    (contoh: total_loss vs iteration).
    """
    if not os.path.exists(metrics_json_path):
        print(f"File {metrics_json_path} tidak ditemukan. Pastikan training telah selesai.")
        return

    iterations = []
    total_loss = []

    with open(metrics_json_path, "r") as f:
        for line in f:
            data = json.loads(line)
            # Hanya tambahkan iteration & total_loss bila keduanya ada.
            if "total_loss" in data:
                iterations.append(data["iteration"])
                total_loss.append(data["total_loss"])

    print(f"Total data di iterations: {len(iterations)}")
    print(f"Total data di total_loss: {len(total_loss)}")

    if len(iterations) == 0:
        print("Tidak ada data total_loss yang ditemukan pada metrics.json.")
        return

    # Plot total_loss vs iteration
    plt.figure(figsize=(8, 5))
    plt.plot(iterations, total_loss, label="Total Loss")
    plt.xlabel("Iteration")
    plt.ylabel("Total Loss")
    plt.title("Training Loss Curve")
    plt.legend()
    plt.grid(True)
    plt.show()


# =========================================
# 2. Buat Confusion Matrix dari Val Set
#    (dengan IoU-based matching)
# =========================================

def create_confusion_matrix(cfg, dataset_name="my_dataset_val", class_names=None, iou_threshold=0.5):
    """
    Menghitung dan menampilkan confusion matrix sederhana untuk object detection
    dengan mencocokkan pred_boxes dan gt_boxes berdasarkan IoU di atas 'iou_threshold'.
    Kelas akan ditampilkan dari dataset (metadata.thing_classes) jika class_names=None.
    """
    cfg.MODEL.WEIGHTS = os.path.join(cfg.OUTPUT_DIR, "model_final.pth")
    predictor = DefaultPredictor(cfg)

    # Ambil data val
    dataset_dicts = DatasetCatalog.get(dataset_name)
    if class_names is None:
        metadata = MetadataCatalog.get(dataset_name)
        class_names = metadata.get("thing_classes", None)  # Daftar nama kelas dataset

    # Untuk menampung label GT & Pred yang sudah di-match
    all_gt_classes = []
    all_pred_classes = []

    for d in dataset_dicts:
        im = cv2.imread(d["file_name"])
        outputs = predictor(im)

        # Ground-truth boxes
        gt_boxes_xyxy = []
        gt_classes = []

        for anno in d["annotations"]:
            # COCO format: [x, y, w, h]
            x, y, w, h = anno["bbox"]
            gt_boxes_xyxy.append([x, y, x + w, y + h])
            gt_classes.append(anno["category_id"])

        if len(gt_boxes_xyxy) == 0:
            # Tidak ada bbox GT => skip
            continue

        # Konversi GT ke tensor CPU [N,4]
        gt_boxes_xyxy = Boxes(torch.tensor(gt_boxes_xyxy, dtype=torch.float32))

        # Prediction boxes => pindahkan ke CPU
        pred_boxes = outputs["instances"].pred_boxes.tensor.cpu()
        pred_classes = outputs["instances"].pred_classes.cpu().numpy()

        if len(pred_boxes) == 0:
            # Tidak ada prediksi => semua GT missed (kita skip atau bisa tambahkan).
            continue

        # Hitung pairwise IoU => keduanya di CPU
        ious = pairwise_iou(gt_boxes_xyxy, Boxes(pred_boxes))

        # Matching IoU (greedy)
        matched_pred_idx = set()
        for i in range(len(gt_boxes_xyxy)):
            iou_row = ious[i]
            best_idx = torch.argmax(iou_row).item()
            best_iou = iou_row[best_idx].item()

            # Jika IoU cukup & pred_box tsb belum dipakai => match
            if best_iou >= iou_threshold and best_idx not in matched_pred_idx:
                matched_pred_idx.add(best_idx)
                all_gt_classes.append(gt_classes[i])
                all_pred_classes.append(pred_classes[best_idx])
            else:
                # GT ini tidak terdeteksi atau IoU < threshold =>
                # Di sini bisa diangap "missed detection",
                # misalnya masukkan "background", dsb (opsional).
                pass

        # (opsional) Tangani false positives di pred_boxes yang tidak terpakai

    # Buat confusion matrix
    from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

    num_classes = cfg.MODEL.ROI_HEADS.NUM_CLASSES
    # Asumsikan label [0..num_classes-1]
    labels = list(range(num_classes))

    cm = confusion_matrix(all_gt_classes, all_pred_classes, labels=labels)

    # Tampilkan
    if class_names is None:
        disp_labels = [str(i) for i in labels]
    else:
        disp_labels = class_names

    cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=disp_labels)
    plt.figure(figsize=(6, 6))
    cm_display.plot(values_format='d', cmap='Blues', colorbar=False)
    plt.title("Confusion Matrix (IoU-based)")
    plt.show()


# =========================================
# 3. Main Eksekusi (Contoh Penggunaan)
# =========================================

if __name__ == "__main__":
  # Gunakan custom backbone MobileNetV3Small dengan FPN
    cfg.MODEL.BACKBONE.NAME = "build_mobilenetv3_small_fpn_backbone"
    cfg.MODEL.FPN.IN_FEATURES = ["p2", "p3", "p4", "p5"]

    # FPN output akan memiliki fitur: p2, p3, p4, p5, dan p6 (dari top block)
    cfg.MODEL.RPN.IN_FEATURES = ["p2", "p3", "p4", "p5", "p6"]
    cfg.MODEL.ROI_HEADS.IN_FEATURES = ["p2", "p3", "p4", "p5", "p6"]

    # Sesuaikan anchor sizes (misal: level p2 mendapatkan anchor terkecil, 16 piksel)
    cfg.MODEL.ANCHOR_GENERATOR.SIZES = [[16], [32], [64], [128], [256]]

    cfg.MODEL.ROI_HEADS.NUM_CLASSES = 4
    cfg.MODEL.WEIGHTS = ""
    cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
    cfg.OUTPUT_DIR = "./output_mbv3_fpn"

    cfg.DATASETS.TRAIN = ("my_dataset_train",)
    cfg.DATASETS.TEST = ("my_dataset_val",)
    cfg.DATALOADER.NUM_WORKERS = 2

    cfg.SOLVER.IMS_PER_BATCH = 2
    cfg.SOLVER.BASE_LR = 0.002
    cfg.SOLVER.MAX_ITER = 20000
    cfg.SOLVER.STEPS = (12000, 16000)
    cfg.SOLVER.GAMMA = 0.1
    cfg.SOLVER.WARMUP_ITERS = 1000
    cfg.SOLVER.WARMUP_FACTOR = 1.0 / 1000
    cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128

    cfg.MODEL.FPN.SQUARE_PAD = 0

    cfg.SOLVER.OPTIMIZER = "SGD"
    cfg.SOLVER.MOMENTUM = 0.9
    cfg.SOLVER.WEIGHT_DECAY = 1e-4



    # 2) Plot grafik training loss
    plot_training_curve(metrics_json_path=os.path.join(cfg.OUTPUT_DIR, "metrics.json"))

    # 3) Tampilkan confusion matrix dgn nama kelas sesuai dataset
    #    => class_names=None agar otomatis ambil metadata
    create_confusion_matrix(cfg, dataset_name="my_dataset_val", class_names=None)
```

## **Hitung Jumlah Parameter**

```python
# ==================
# 1. Import & Setup
# ==================
import torch
import torch.nn as nn

from detectron2.config import get_cfg
from detectron2.engine import DefaultTrainer
from detectron2.modeling import BACKBONE_REGISTRY, Backbone, build_backbone, build_model
from detectron2.layers import Conv2d, ShapeSpec
# Import FPN & LastLevelMaxPool dari submodule fpn
from detectron2.modeling.backbone.fpn import FPN, LastLevelMaxPool

from detectron2.checkpoint import DetectionCheckpointer
from detectron2.data import build_detection_test_loader
from detectron2.evaluation import inference_on_dataset, COCOEvaluator
from detectron2.utils.logger import setup_logger

setup_logger()

# ==========================================
# 2. Registrasi Dataset (Opsional)
# ==========================================
# Jika diperlukan, Anda bisa mendaftarkan dataset Anda. Contoh:
# from detectron2.data.datasets import register_coco_instances
# register_coco_instances("my_dataset_train", {}, "path/to/train.json", "path/to/images_train")
# register_coco_instances("my_dataset_val", {}, "path/to/val.json", "path/to/images_val")

# ======================================================
# 3. Definisikan MobileNetV3 Small sebagai custom backbone (bottom-up)
# ======================================================
@BACKBONE_REGISTRY.register()
def build_mobilenetv3_small_backbone(cfg, input_shape: ShapeSpec):
    import torchvision.models as models

    # Gunakan pre-trained weights untuk MobileNetV3 Small
    weights = models.MobileNet_V3_Small_Weights.IMAGENET1K_V1
    model = models.mobilenet_v3_small(weights=weights)

    # Pembagian stage sesuai downsampling yang diharapkan:
    # stage1: features[:2]  -> 16 channel, effective stride ~4
    # stage2: features[2:4]  -> 24 channel, effective stride ~8
    # stage3: features[4:6]  -> 40 channel, effective stride ~16
    # stage4: features[6:9]  -> 48 channel, tetap dengan stride ~16
    # stage5: features[9:]   -> 576 channel, effective stride 32
    stage1 = nn.Sequential(*model.features[:2])
    stage2 = nn.Sequential(*model.features[2:4])
    stage3 = nn.Sequential(*model.features[4:6])
    stage4 = nn.Sequential(*model.features[6:9])
    stage5 = nn.Sequential(*model.features[9:])

    class MobileNetV3Backbone(Backbone):
        def __init__(self):
            super().__init__()
            self.stage1 = stage1
            self.stage2 = stage2
            self.stage3 = stage3
            self.stage4 = stage4
            self.stage5 = stage5

            # (Opsional) Modul peningkatan untuk fitur resolusi tinggi (dari stage1)
            self.extra_p2 = nn.Sequential(
                Conv2d(16, 16, kernel_size=3, padding=1, bias=False),
                nn.BatchNorm2d(16),
                nn.ReLU()
            )

            # Proyeksi ke 256 channel untuk FPN:
            # p2: dari stage1 (stride 4)
            # p3: dari stage2 (stride 8)
            # p4: dari stage3 & stage4 (stride 16)
            # p5: dari stage5 (stride 32)
            self.conv_p2 = Conv2d(16, 256, kernel_size=1, bias=False)
            self.conv_p3 = Conv2d(24, 256, kernel_size=1, bias=False)
            self.conv_p4 = Conv2d(48, 256, kernel_size=1, bias=False)
            self.conv_p5 = Conv2d(576, 256, kernel_size=1, bias=False)

        def forward(self, x):
            # Stage 1: Output stride ~4
            x = self.stage1(x)
            x_enhanced = self.extra_p2(x)
            p2 = self.conv_p2(x_enhanced)

            # Stage 2: Output stride ~8
            x = self.stage2(x)
            p3 = self.conv_p3(x)

            # Stage 3 & 4:
            # Stage3 menghasilkan 40 channel dengan stride ~16, kemudian diproses di stage4 untuk menghasilkan 48 channel
            x = self.stage3(x)
            x = self.stage4(x)
            p4 = self.conv_p4(x)

            # Stage 5: Output stride 32
            x = self.stage5(x)
            p5 = self.conv_p5(x)

            return {"p2": p2, "p3": p3, "p4": p4, "p5": p5}

        def output_shape(self):
            return {
                "p2": ShapeSpec(channels=256, stride=4),
                "p3": ShapeSpec(channels=256, stride=8),
                "p4": ShapeSpec(channels=256, stride=16),
                "p5": ShapeSpec(channels=256, stride=32),
            }

    return MobileNetV3Backbone()

# ======================================================
# 4. Bungkus backbone di atas dengan FPN
# ======================================================
@BACKBONE_REGISTRY.register()
def build_mobilenetv3_small_fpn_backbone(cfg, input_shape):
    """
    Membungkus MobileNetV3Small backbone dengan FPN.
    Menggunakan output fitur: p2, p3, p4, p5 → FPN akan menghasilkan p2, p3, p4, p5, dan top block menghasilkan p6.
    """
    bottom_up = build_mobilenetv3_small_backbone(cfg, input_shape)
    fpn = FPN(
        bottom_up=bottom_up,
        in_features=["p2", "p3", "p4", "p5"],
        out_channels=256,
        norm=cfg.MODEL.FPN.NORM,
        top_block=LastLevelMaxPool(),  # Top block akan menghasilkan fitur "p6"
        square_pad=cfg.MODEL.FPN.SQUARE_PAD
    )
    return fpn

# ==========================================
# 5. Configurasi
# ==========================================
cfg = get_cfg()
cfg.merge_from_file("detectron2/configs/COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml")

# Gunakan custom backbone MobileNetV3Small dengan FPN
cfg.MODEL.BACKBONE.NAME = "build_mobilenetv3_small_fpn_backbone"
cfg.MODEL.FPN.IN_FEATURES = ["p2", "p3", "p4", "p5"]

# FPN output akan memiliki fitur: p2, p3, p4, p5, dan p6 (dari top block)
cfg.MODEL.RPN.IN_FEATURES = ["p2", "p3", "p4", "p5", "p6"]
cfg.MODEL.ROI_HEADS.IN_FEATURES = ["p2", "p3", "p4", "p5", "p6"]

# Sesuaikan anchor sizes (misal: level p2 mendapatkan anchor terkecil, 16 piksel)
cfg.MODEL.ANCHOR_GENERATOR.SIZES = [[16], [32], [64], [128], [256]]

cfg.MODEL.ROI_HEADS.NUM_CLASSES = 4
cfg.MODEL.WEIGHTS = ""
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
cfg.OUTPUT_DIR = "./output_mbv3_fpn"

cfg.DATASETS.TRAIN = ("my_dataset_train",)
cfg.DATASETS.TEST = ("my_dataset_val",)
cfg.DATALOADER.NUM_WORKERS = 2

cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.002
cfg.SOLVER.MAX_ITER = 20000
cfg.SOLVER.STEPS = (12000, 16000)
cfg.SOLVER.GAMMA = 0.1
cfg.SOLVER.WARMUP_ITERS = 1000
cfg.SOLVER.WARMUP_FACTOR = 1.0 / 1000
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128

cfg.MODEL.FPN.SQUARE_PAD = 0

cfg.SOLVER.OPTIMIZER = "SGD"
cfg.SOLVER.MOMENTUM = 0.9
cfg.SOLVER.WEIGHT_DECAY = 1e-4

# =============================
# 2. Bangun model dan hitung parameter
# =============================

# Bangun model berdasarkan konfigurasi di atas
model = build_model(cfg)
model.eval()  # Set model ke mode evaluasi (tidak diperlukan grad)

# Hitung total parameter
total_params = sum(p.numel() for p in model.parameters())
# Hitung parameter yang dapat di-train (memiliki flag requires_grad=True)
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print("Total jumlah parameter: ", total_params)
print("Jumlah parameter yang trainable: ", trainable_params)
```

```python
import time
import torch
import cv2
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg

# Pastikan konfigurasi dan custom backbone sudah didefinisikan sebelumnya

# Load model yang sudah dilatih (sesuaikan path jika diperlukan)
cfg.MODEL.WEIGHTS = "/content/drive/MyDrive/KULIAH/SKRIPSI/FIX_MODIFIKASI/output_mbv3_fpn_3/model_final.pth"

# Buat objek predictor
predictor = DefaultPredictor(cfg)

# Baca gambar uji (pastikan resolusi dan preprocessing sesuai)
image = cv2.imread("/content/WhatsApp Image 2025-01-31 at 12.23.36_cb3a5c65.jpg")

num_iter = 100  # jumlah iterasi per pengukuran
num_repeat = 10  # jumlah pengukuran
fps_list = []

for i in range(num_repeat):
    # Lakukan warming-up setiap pengukuran agar mendapatkan hasil yang konsisten
    for _ in range(10):
        _ = predictor(image)
        if torch.cuda.is_available():
            torch.cuda.synchronize()

    # Pengukuran inferensi
    start_time = time.time()
    for _ in range(num_iter):
        _ = predictor(image)
        if torch.cuda.is_available():
            torch.cuda.synchronize()
    end_time = time.time()

    total_time = end_time - start_time
    fps = num_iter / total_time
    print(f"Run {i+1} FPS: {fps:.2f}")
    fps_list.append(fps)

# Hitung rata-rata FPS dari semua pengukuran
avg_fps = sum(fps_list) / len(fps_list)
print(f"Average FPS over {num_repeat} runs: {avg_fps:.2f}")
```
